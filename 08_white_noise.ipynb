# White Noise

Requirements:
- Independent variables
- Variables are identically distributed
- All means are equal to zero

Therefore:
- All variables have the same variance
- Each value has zero correlation with all other values in the series

If a time series is white noise, it is random. It cannot be reasonable modeled nor can predictions be made. 

The errors from a time series forecast model should ideally be white noise.

### Questions:
1. Does the series have a non-zero mean?
2. Does the variance change over time?
3. Do values correlate with lag values?

You can use the following tools to evaluate these questions:
- Line plot
- Summary statistics
- Autocorrelation plot

We can create a white noise series to use as a reference for other time series sets:

```python
import random
import pandas as pd
from pandas.plotting import autocorrelation_plot
import matplotlib.pyplot as plt
```

```python
# random number seed
random.seed(1)

# create white noise series
series = [random.gauss(0.0, 1.0) for i in range(1000)]
series = pd.Series(series)
series
```

```python
# Summary statistics
series.describe()
```

The mean is nearly 0 and the standard deviation is nearly 1. 

```python
# line plot
series.plot()
plt.show()
```

```python
# histogram plot
series.hist(edgecolor='black')
plt.show()
```

```python
series.plot(kind='kde')
plt.show()
```

The distribution is a little pointed, but overall still represents the bell shape that we are looking for.

```python
# autocorrelation
autocorrelation_plot(series)
plt.show()
```

The few spikes around statistical significance are flukes (we know this with certainty because the production of this data was random). Overall, there is no pattern of autocorrelation present.

### Why is this important?

Identifying if a series is white noise in nature will allow us to quickly decide if it warrants further analysis and modeling. The more a series approaches white noise, the more difficult it will be to create a model that provides value.
